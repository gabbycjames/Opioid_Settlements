## Visualizations

```python
row = df.iloc[0, 5:]
```

```python
print(row)
```

```python
plt.plot(row)
plt.xlabel('Year (Actual or Expected)')
plt.ylabel('Payout in $ (1,000,000)')
plt.title('Payout Over Time for St. Louis County')
plt.xticks(rotation=45)
plt.show()
```

```python
stlcity = df.iloc[4, 5:]
plt.plot(stlcity)
plt.xlabel('Year (Actual or Expected)')
plt.ylabel('Payout in $ (1,000,000)')
plt.title('Payout Over Time for St. Louis City')
plt.xticks(rotation=45)
plt.show()
```

```python
stl = df.iloc[[0, 4], 5:]
stl_transposed = stl.T
```

```python
location = {'0':'St. Louis Country', '4':'St. Louis City'}
labels = list(location.keys())
labels = [location[key] for key in labels]
stl_transposed.plot()
plt.xlabel('Year (Actual or Expected)')
plt.ylabel('Payout in $ (1,000,000)')
plt.title('Payout Over Time for St. Louis City vs St. Louis County')
plt.legend(labels)
plt.xticks(rotation=45)
plt.show()
```
