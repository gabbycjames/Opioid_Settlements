## Visualizations

```python
df.hist(figsize = (25,20))
plt.subplots_adjust(hspace=0.5);
```

```python
corr = df.corr(numeric_only=True)
corr
```

```python
#plt.figure(figsize=(20,20))
_ = sns.heatmap(corr, annot=False)
plt.show()
```

```python
plt.figure(figsize=(40,40))
plt.bar(df['beneficiary_name'],df['total_paid'])
plt.xlabel('Beneficiary/County')
plt.ylabel('Payout in $ (1,000,000)')
plt.title('Total Paid by County')
plt.xticks(rotation=45)
plt.show()
```

```python
df['total_paid'].sort_values(ascending=False)
```

```python
filtered_data = df[df['total_paid'] > 100000]

# Plot the filtered data using Seaborn
plt.figure(figsize=(20, 5))
sns.barplot(x=filtered_data['county'], y=filtered_data['total_paid'])
plt.xticks(rotation=90)
plt.xlabel('County')
plt.ylabel('Values greater than $100,000')
plt.title('Bar Plot of Values Greater Than 0')
plt.show()
```

```python
print(row)
```

```python
plt.plot(row)
plt.xlabel('Year (Actual or Expected)')
plt.ylabel('Payout in $ (1,000,000)')
plt.title('Payout Over Time for St. Louis County')
plt.xticks(rotation=45)
plt.show()
```

```python
stlcity = df.iloc[4, 5:]
plt.plot(stlcity)
plt.xlabel('Year (Actual or Expected)')
plt.ylabel('Payout in $ (1,000,000)')
plt.title('Payout Over Time for St. Louis City')
plt.xticks(rotation=45)
plt.show()
```

```python
stl = df.iloc[[0, 4], 5:]
stl_transposed = stl.T
```

```python
location = {'0':'St. Louis County', '4':'St. Louis City'}
labels = list(location.keys())
labels = [location[key] for key in labels]
stl_transposed.plot()
plt.xlabel('Year (Actual or Expected)')
plt.ylabel('Payout in $ (1,000,000)')
plt.title('Payout Over Time for St. Louis City vs St. Louis County')
plt.legend(labels)
plt.xticks(rotation=45)
plt.show()
```

```python
stl_transposed.plot(kind='bar')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
years = ['paid_2022', 'paid_2023', 'est_2024', 'est_2025', 'est_2026', 'est_2027', 'est_2028', 'est_2029', 'est_2030', 'est_2031', 'est_2032']
df[years].sum().plot(kind='line', marker='o')
plt.title('Total Payments by Year (2022-2032)')
plt.xlabel('Year')
plt.ylabel('Amount Paid')
plt.show()
```

```python
plt.figure(figsize=(20,20))
sns.heatmap(df.pivot_table(index='county', values=years, aggfunc='sum'), cmap='Blues', annot=True)
plt.xticks(fontsize=5)
plt.yticks(fontsize=5)

plt.title('Yearly Payments by County')
plt.show()
```

```python
plt.figure(figsize=(20, 20))
df.groupby('beneficiary_name')['total_paid'].sum().plot(kind='pie', startangle=90)
plt.legend(loc='upper left', bbox_to_anchor=(1, 0, 0.5, 1))
plt.title('Settlement Allocation by Beneficiary')
plt.ylabel('')
plt.show()
```
